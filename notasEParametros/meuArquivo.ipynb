```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
df["parental level of education"].value_counts()/10
## porcentagens de cada nível de educação dos pais
```

```python
df.groupby("parental level of education")["math score"].mean()
## média da pontuação em matemática por nível de educação dos pais
```

```python
df.groupby("parental level of education")[["math score", "writing score", "reading score"]].mean()/10
## média das pontuações em matemática, escrita e leitura por nível de educação dos pais
```

```python
df.groupby(["parental level of education","race/ethnicity"])[["math score", "writing score", "reading score"]].mean()
## média das pontuações em matemática, escrita e leitura por nível de educação dos pais e raça/etnia
```

```python
df["gender"].value_counts()
```

```python
# Normalização por proporção (percentuais)
proporcoes_genero = df["gender"].value_counts(normalize=True) * 100
print("Proporções em %:")
print(proporcoes_genero)

# Ou usar value_counts diretamente
print("\nProporções (0-1):")
print(df["gender"].value_counts(normalize=True))
```

```python
##MEDIA DE MATEMÁTICA POR GENERO

df.groupby("gender")["math score"].mean().reset_index()
```

```python
#MEDIA POR GRUPO ETNICO

df.groupby("race/ethnicity")["math score"].mean().reset_index()
```

```python
df.head(5)
```

```python
#impacto do curso preparatorio

df.groupby("test preparation course")[["math score","reading score","writing score"]].mean().reset_index()
```

```python
#agrupamento por 2 colunas

df.groupby(["gender","test preparation course"])["math score"].mean().reset_index()
```

```python
#varias estatísticas

df.groupby(["gender","test preparation course"])["math score"].agg(["mean","std","min","max"]).reset_index()
```

```python
df.groupby("gender")[["math score","reading score","writing score"]].min().reset_index()
```
